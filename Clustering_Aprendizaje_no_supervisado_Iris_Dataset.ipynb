# **Cargamos librerías**

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

# **Leemos el csv**

```python
df = pd.read_csv('IRIS.csv')
df.head(10)
```

# **Comprobamos datos vacíos**

```python
print(df.isnull().sum())
```

# **Definimos variables para entrenar el modelo (ignoramos a species por obvias razones)**

```python
x = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
scaler = StandardScaler()
X_std = scaler.fit_transform(x)
```

```python
modelo = KMeans(n_clusters=3)
modelo.fit(X_std)
```

# **Agregamos la columna cluster para ver los resultados y comparar**

```python
df['clusters'] = modelo.labels_

df.head(60)
```

# **Comprobaciones de aciertos del modelo entrenado**

```python
contando = df.groupby(['species', 'clusters']).size()

display(contando)
```

```python
clusteres = df.groupby('clusters').size()
clustersPorcentaje = clusteres.div(clusteres, level='clusters') * 100

print(clustersPorcentaje)
```
